## Folder structure
- **data**
    contains input data
    
`/Grid_NetCDF`: grid input file (NetCDF file)

`/RichardsMeshGen_input`: .csv file containing the data required to define the grid. These files are input file for notebooks `1_RichardsMeshGen_xxx_SWRC`

`/Timeseries`: .csv file containing the timeseries used for the boundary conditions. These files have to be formatted as:

    @T,table
    Created,2016-11-08 19:32
    Author,HortonMachine library
    @H,timestamp,value_1
    ID,,1
    Type,Date,Double
    Format,yyyy-MM-dd HH:mm,
    ,yyyy-mm-dd hh:mm,value
    
- **Jupyter_Notebook**
    contains the notebook for pre-processing and post-processing

- **lib**
    contains .jar file to run the simulations
    
- **docs**
    contains a brief documentation of this documentation
    
- **output**
    contains the output of the simulations
    
- **simulation**
    contains the `.sim` file
    
## Notebooks

The python scripts used in the notebooks are contained in some python files that are imported in the notebook running these lines:
    
`%%capture`
    
`%run file_name.py`

This file has to be in the same folder of the notebook and you can import only one file per notebook.

### Notebook contents

- **0_How_to_Read_NetCDF**
    A smooth introduction to NetCDF file format and to handle them
    
- **1_RichardsMeshGen_BrooksCorey_SWRC**
    Notebook to create the input file for the grid when the Brooks and Corey SWRC model is choosen. Functions are defined in `RichardsMeshGenBrooksCorey.py`.
    
- **1_RichardsMeshGen_Kosugi_SWRC**
    Notebook to create the input file for the grid when the Kosugi SWRC model is choosen. Functions are defined in `RichardsMeshGenKosugi.py`.

- **1_RichardsMeshGen_VanGenuchten_SWRC**
    Notebook to create the input file for the grid when the Van Genuchten SWRC model is choosen. Functions are defined in `RichardsMeshGenVanGenuchten.py`.
    
- **1_RichardsMeshGen_Romano_SWRC**
    Notebook to create the input file for the grid when the Romano SWRC model is choosen. Functions are defined in `RichardsMeshGenRomano.py`.
        
- **2_Run_Docker**
    Notebook to run the .sim with Docker.
    
- **3_Richards1DOutput**
   Notebook to visualize the output of the simulation. The scripts to plot data are defined as function in `Richards1DOutput.py`.

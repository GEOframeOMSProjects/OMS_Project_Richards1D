Per risolvere un'equazione alle derivate parziali (PDE) è necessario definire:
- un'equazione di stato, nel caso dell'equazione di Richards è rappresentata dalle curve di ritenzione idrica (SWRC);
- un'equazione di chiusura, nel caso dell' equazione di Richards è il modello di Mualem che stabilisce come varia la conducibilità idraulica con il grado di saturazione del suolo (esistono anche altri modelli per definire la conducibilità idraulica in un mezzo insaturo);
- una condizione iniziale;
- le condizioni al contorno.

Da un punto di vista fisico la **condizione iniziale** rappresenta lo stato iniziale del sistema. Dal punto di vista matematico è una funzione del tipo:
$$\psi(z,0) = \psi_0(z)$$
ovvero una funzione che dipende esclusivamente dello spazio.

Per tempi vicini a $t_0$ lo stato del sistema dipende in modo significativo dalla condizione iniziale, per tempi lunghi questa dipendenza diventa via via meno significativa. In questo caso si dice che il sistema perde memoria della condizione iniziale.

_Come viene assengata la condizione iniziale?_
- misure di suzione;
- ipotizzata. In questo caso si può assumere che l'acuqa nel suolo si in equilibrio (condizione idrostatica) e si ipotizza un certa profondità della falda, che in altre parole significa stabilire a che profondità si abbia $\psi=0$.
- data l'incertezza nel determinare la condizione iniziale, si può fare uno _spin-up_ del modello. Viene ipotizzata una condizione iniziale, viene eseguita una simulazione di 1-2 anni a seconda della profondità del terreno con delle condizioni al contorno misurate o simulate (ad esempio per la precipitazione si possono utilizzare serie storiche oppure delle serie sintetiche ) e la soluzione che si ottiene rappresenta la condizione iniziale per il problema in esame. Questa tecnica si basa sul fatto che il sistema nel tempo perde memoria della sua condizione iniziale e il suo stato è determinato dalle sole condizioni al contorno.

## Plot results
The solution is time dependent,thus to show the solution at a certain time you can define the timeIndex variable (i.e. the number of the time step: 0 is the first time step coincedent with the the start date of your simulation, 1 is the second one and so on), or define myDate (i.e. day and time you are interested in) 

```python
## Enter the dates [yyyy-MM-dd HH:mm] you want to plot
myDate = ['2017-01-01 01:00', '2017-01-07 00:00', '2017-01-14 00:00','2017-01-21 00:00','2017-01-28 00:00']
timeIndex = []
for i in range(0,len(myDate)):
    timeIndex = np.append(timeIndex,datesHuman.index(myDate[i]))
date = myDate
for d in date:
    print('Considero la soluzione per '+ d) 
```

### Plot initial condition

```python
showInitialCondition(iC,depths,ncfile,labelSize,titleSize,legendSize,axisTicksSize,lineWidth,lineStyle,markerSize,markerType,figureSizeHeigth,figureSizeWidth, 'upper left')
```

In questo caso la condizione iniziale non è una condzione di equilibrio. Infatti, i flussi dipendono dal gradiente del carico idraulico $h$ che come si vede non è costante lungo la verticale. L'acqua si muove dall'alto verso il basso. Fissata la coordinata $z$ positiva verso l'alto, la legge di Darcy stabilisce che il moto dell'acqua avviente nel verso rispetto al gradiente di $h$ che in questo che in questo caso è positivo.

### Plot water suction

```python
showWaterSuction(timeIndex,date,psi,depths,time,ncfile,labelSize,titleSize,legendSize,axisTicksSize,lineWidth,lineStyle,markerSize,markerType,figureSizeHeigth,figureSizeWidth,'lower left')
```

In accordo a quanto osservato prima, dal grafico si osserva come negli strati superficiali la suzione diminuisce mentre in quelli inferiori aumenta. In questo caso l'evoluzione temporale è dovuta non alle condizioni al contorno (flusso nullo) ma allo squilibrio della condione iniziale del sistema. Si osserva che il sistema tende a portarsi verso la condizione idrostatica ovvero alla condizione di equilibrio.

Il carico idraulico tende a portarsi verso la verticale
